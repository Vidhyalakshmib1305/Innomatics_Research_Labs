### 3.1 Outlier Detection For Numerical Columns.

Outlier Detection For
(i) Salary
(ii) 10percentage
(iii) Domain

```python
# (i) Salary

# Step 1: Calculate Interquartile Range (IQR)
Q1 = df1['Salary'].quantile(0.25)
Q3 = df1['Salary'].quantile(0.75)
IQR = Q3 - Q1

# Step 2: Identify Outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = df1[(df1['Salary'] < lower_bound) | (df1['Salary'] > upper_bound)]

# Step 3: Visualize Outliers
plt.figure(figsize=(8, 6))
sns.boxplot(df1['Salary'])
plt.xlabel('Salary')
plt.title('Boxplot of Salary with Outliers')
plt.scatter(x=outliers.index, y=outliers['Salary'], color='red', label='Outliers')
plt.legend()
plt.show()

# Step 4: Consider Context and Domain Knowledge
# Analyze the outliers in the context of the data and domain knowledge to determine their significance and potential causes.
```

### OBSERVATION

key observations I can make:

1. Salary distribution: The boxplot shows that the salaries in the dataset are not normally distributed. The median salary is around $150,000, but there is a long tail of outliers towards the higher end of the salary range. This means that a few people earn much more than the majority of the people in the dataset.

2. Interquartile range: The interquartile range (IQR) is the difference between the 75th and 25th percentile. In this boxplot, the IQR is around $75,000.
This means that the middle 50% of the people in the dataset earn between $ 75,000 and $ 150,000.

3. Outliers: There are several outliers in the data, both above and below the box. Outliers are data points that fall outside of the 1.5 IQR range. The presence of outliers can significantly impact the interpretation of statistical measures such as the mean and standard deviation.

## Identifying Relationships between categorical and categorical columns

```python
# Select categorical columns
categorical_columns = ['Gender', 'Designation']  # Example

# Get the top N most frequent designations
top_designations = df1['Designation'].value_counts().nlargest(10).index  # Adjust 10 as needed

# Filter data to include only top designations
filtered_data = df1[df1['Designation'].isin(top_designations)]

# Create cross-tabulation
cross_tab = pd.crosstab(filtered_data[categorical_columns[0]], filtered_data[categorical_columns[1]])

# Plot stacked bar plot
cross_tab.plot(kind='bar', stacked=True)
plt.title('Stacked Bar Plot of {} by {}'.format(categorical_columns[1], categorical_columns[0]))
plt.xlabel(categorical_columns[0])
plt.ylabel('Count')
plt.xticks(rotation=0)  # Rotate x-axis labels if necessary
plt.tight_layout()  # Adjust layout to prevent overlapping
plt.show()
```

### OBSERVATION

The x-axis represents the gender.
The y-axis represents the count of employees.
Each bar represents a designation, and the segments within the bar are colored blue and orange to represent the count of male and female employees, respectively.
